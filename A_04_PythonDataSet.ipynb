# Barchart

```python

pylab.bar(df['innings1_runs'],df['innings2_runs'],color='red')
pylab.xlabel('innings1_runs')
pylab.ylabel('innings2_runs')
pylab.show()
```

# Horizontal

```python

pylab.barh(df['innings1_runs'],df['innings2_runs'],color='red')
pylab.xlabel('innings1_runs')
pylab.ylabel('innings2_runs')
pylab.show()
```

# Pie chart

```python
data_pie = df['venue'].value_counts().rename_axis('venue').reset_index(name='patients_count')
pylab.figure(figsize=(10,10))
pylab.pie(data_pie.patients_count,labels=data_pie.venue,startangle=90,autopct='%.1f%%')
pylab.title("The Venus")
pylab.show()
```
